```python
# separates movie genres by comma
actors = df['Actors'].str.split(',')

# we need to remove spaces before or after the name
actors = actors.apply(lambda x: [i.strip() for i in x])  

# explode method puts each type on a separate line
actors = actors.explode()

# the top 20 actors according to the number of movies they starred in
actors.value_counts()
```

```python
plt.figure(facecolor = "Violet",figsize = (12, 6))
ax= sns.barplot(y = actors.value_counts().head(20).index, x = actors.value_counts().head(20).values)

ax.set(xlabel='Count',ylabel='Actors', title='Top 20 Actors by Number of Movies')
ax.set_facecolor("xkcd:mint green")
ax.bar_label(ax.containers[0])

plt.show()
```

```python
# average rating scores of the top 20 actors with the most films
runtime = []
for i in actors.value_counts().index:
    x = df[df.Actors.str.contains(i)]["Runtime (Minutes)"].mean()
    runtime.append(x)

rating = []
for i in actors.value_counts().index:
    x = df[df.Actors.str.contains(i)]["Rating"].mean()
    rating.append(x)
    
actors_run_rat = pd.DataFrame({'Runtime': runtime, "Rating": rating}, index=actors.value_counts().index)
actors_run_rat
```

```python
sns.scatterplot(data=actors_run_rat, x="Rating", y="Runtime");
```

```python
df.Director.value_counts()[:10]
director_10 = df[df.Director.isin(df.Director.value_counts()[:10].index)]
director_10.head(2)
```

```python
director_10.groupby("Director")["Revenue (Millions)"].mean()
```

```python
fig, ax1 = plt.subplots(figsize = (14,8))
plt.xticks(rotation=45)
ax2 = ax1.twinx()     

ax1.plot(director_10.groupby("Director").size().index, director_10.groupby("Director").size().values, color='blue', label='Num. of Movies')
ax1.set_ylabel("Number of Movies",fontsize = 12, color = "b")


ax2.plot(director_10.groupby("Director")["Revenue (Millions)"].mean().index, director_10.groupby("Director")["Revenue (Millions)"].mean(), color='red', label='Revenue')
ax2.set_ylabel('Average Revenue', color = "r",fontsize = 12)

ax2.spines['right'].set_color('red')
ax2.spines["left"].set_color("blue")

ax1.set_xlabel('Director',fontsize = 12)
# ax1.set_title('Yield Curve Comparison: July 16, 2007 vs. July 16, 2020')  ax1 veya ax2 üzerinden ortak set_title çalışır
ax2.set_title('The number of films and revenue averages for the directors with the most number of movies(Top10)')

ax1.legend(loc='upper left')
ax2.legend(loc='upper center')

plt.show()
```
